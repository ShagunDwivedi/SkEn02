```python
df = pd.read_csv('/kaggle/input/supermarket-sales/supermarket_sales - Sheet1.csv')
df['Date'] = pd.to_datetime(df.Date).dt.strftime('%Y-%m-%d')
df.head()
```

```python
df[['Branch','City']].value_counts()
```

```python
pv_df = pd.pivot_table(data=df,index=['City','Product line'],columns='Date', values='Total', aggfunc='sum')
pv_df.fillna(0, inplace=True)
pv_df = pv_df.reset_index()
pv_df.head()
```

```python
for i in range(3,len(pv_df.columns)):
    pv_df[pv_df.columns[i]] += pv_df[pv_df.columns[i-1]]
```
