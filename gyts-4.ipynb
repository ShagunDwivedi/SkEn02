```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
dat = pd.read_csv('assets/datafile.csv')
dat.head()
```

```python
dat['State/UT'].unique()
```

```python
dat[dat['State/UT']=='West Bengal']
#there's no urban in this state for some reason????
```

```python
dat.columns
```

```python
lu = dat[dat['Area']=='Urban']['Current cigarette users (%)']
z = dat[dat['Area']=='Urban']['State/UT']
lr = dat[dat['Area']=='Total']['Current cigarette users (%)']
```

```python
fig, ax = plt.subplots()
x = np.arange(len(z))
width = 0.3

rects1 = ax.bar(x - width/2, lu, width, label='Urban')
rects2 = ax.bar(x + width/2, lr, width, label='Total')

ax.set_ylabel('% of users')
ax.set_ylim(0,70)
ax.set_title('Cigarette Usage by Urban/Total')
ax.set_xticks(x, z)
ax.legend(loc='best', fontsize=8)

#ax.bar_label(rects1, padding=3, fontsize=6)
#ax.bar_label(rects2, padding=3, fontsize=6)
plt.xticks(fontsize=8, rotation=90)
plt.grid(axis='y', linestyle='dotted')

fig.tight_layout()

plt.show()
```

```python
lr = dat[dat['Area']=='Urban']['Current tobacco users (%)']
```

```python
fig, ax = plt.subplots()
x = np.arange(len(z))
width = 0.3

rects1 = ax.bar(x - width/2, lu, width, label='Cig')
rects2 = ax.bar(x + width/2, lr, width, label='Tob')

ax.set_ylabel('% of users')
ax.set_ylim(0,65)
ax.set_title('Cigarette vs Tobacco Usage in Urban Indian Youth')
ax.set_xticks(x, z)
ax.legend(loc='best', fontsize=8)

#ax.bar_label(rects1, padding=3, fontsize=6)
#ax.bar_label(rects2, padding=3, fontsize=6)
plt.xticks(fontsize=8, rotation=90)
plt.grid(axis='y', linestyle='dotted')

fig.tight_layout()

plt.show()
```
